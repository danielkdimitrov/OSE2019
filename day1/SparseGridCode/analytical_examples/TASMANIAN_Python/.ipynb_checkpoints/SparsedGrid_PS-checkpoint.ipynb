# Problem No. 1 (Analytical Examples)

```python
import TasmanianSG
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


# imports specifically needed by the examples
import math
from random import uniform
from datetime import datetime

#from mpl_toolkits.mplot3d import Axes3D

print("TasmanianSG version: {0:s}".format(TasmanianSG.__version__))
print("TasmanianSG license: {0:s}".format(TasmanianSG.__license__))

grid  = TasmanianSG.TasmanianSparseGrid()
grid1 = TasmanianSG.TasmanianSparseGrid()
grid2 = TasmanianSG.TasmanianSparseGrid()
```

I use the functions

+ function 1 : $f(x,y) = cos(2\pi w_1 + c_1x + c_2y)$ 

+ function 2: $f(x,y) = ((c_1^{-2}+(x-w_1)^2)*(c_1^{-2}+(x-w_2)^2))^{-1}$ 

```python
def test_fn(x,y,type):
    w_1 = .5
    w_2 = .1
    c_1 = 1/3
    c_2 = 1/5
    if type ==1:
        f = np.cos(2*np.pi*w_1 + c_1*x + c_2*y)
    elif type ==2:
        f = ((c_1**(-2)+(x-w_1)**2)*(c_1**(-2)+(x-w_2)**2))**(-1)
    return f
```

```python
#############################################################################

# EXAMPLE 1 for OSM:
# interpolate: fn_test()  
# using piecewise linear basis functions.

def random_grid(fn,fn_type=1):
    # 1000 2-dimensional sample points 
    aPnts = np.empty([1000, 2])  #non-sparse grid test points
    for iI in range(1000):
        for iJ in range(2):
            aPnts[iI][iJ] = uniform(-1.0, 1.0) ## why random ? 

    # Result
    aTres = np.empty([1000,])
    for iI in range(1000):
        aTres[iI] = fn(aPnts[iI][0],aPnts[iI][1],fn_type)
    return aPnts, aTres
    
# Sparse Grid with dimension 2 and 1 output and refinement level 5

def sparse_grid(fn, fn_type, iDepth):
    iDim = 2
    iOut = 1
    #iDepth = 7 # depth of sparse grid, number of points increasing in the depth 
    which_basis = 1 #1= linear basis functions -> Check the manual for other options

    # construct sparse grid
    grid.makeLocalPolynomialGrid(iDim, iOut, iDepth, which_basis, "localp")
    aPoints = grid.getPoints() #x and y, sparse grid  

    iNumP1 = aPoints.shape[0]
    aVals = np.empty([aPoints.shape[0], 1])
    for iI in range(aPoints.shape[0]):
        aVals[iI] = test_fn(aPoints[iI][0],aPoints[iI][1],fn_type)
        
    #grid.plotPoints2D()
    return aPoints, aVals
```

```python
fn_type = 1

aPnts, aTres = random_grid(test_fn,fn_type)

aPoints, aVals = sparse_grid(test_fn,fn_type,5)


fig = plt.figure()
ax = fig.add_subplot(221,projection='3d')
ax.scatter(aPnts[:,0],aPnts[:,1],aTres[:],c='r',marker='o')
plt.title('Random Grid, Fn 1')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(222,projection='3d')
ax.scatter(aPoints[:,0],aPoints[:,1],aVals[:],c='r',marker='o')
plt.title('Sparse Grid, Fn 1')
plt.show()



fn_type = 2

aPnts, aTres = random_grid(test_fn,fn_type)

aPoints, aVals = sparse_grid(test_fn,fn_type,5)

fig = plt.figure()
ax = fig.add_subplot(223,projection='3d')
ax.scatter(aPnts[:,0],aPnts[:,1],aTres[:],c='r',marker='o')
plt.title('Random Grid, Fn 2')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(224,projection='3d')
ax.scatter(aPoints[:,0],aPoints[:,1],aVals[:],c='r',marker='o')
plt.title('Sparse Grid, Fn 2')
plt.show()
```

```python
#############################################################################

## EXAMPLE 2 for OSM:
## interpolate: f(x) = cos(2* pi * w1 + sum(c_i*x_i)) 
## using refinement

def adaptive_sparse_grid(fn, fn_type, iDepth, refinement_level):
    # 1000 2-dimensional sample points 
    aPnts = np.empty([1000, 2])  #non-sparse grid test points
    for iI in range(1000):
        for iJ in range(2):
            aPnts[iI][iJ] = uniform(-1.0, 1.0) ## why random ? 

    aTres = np.empty([1000,])
    for iI in range(1000):
        aTres[iI] = fn(aPnts[iI][0],aPnts[iI][1],fn_type) 
        #math.cos(2 * math.pi * w1 + c_1*aPnts[iI][0] + c_2*aPnts[iI][1]) 

    # Adaptive Sparse Grid with dimension 2 and 1 output and maximum refinement level 5, refinement criterion.
    iDim = 2
    iOut = 1
    #iDepth = 1
    fTol = 1.E-5
    which_basis = 1 
    #refinement_level = 5

    # level of grid before refinement
    grid1.makeLocalPolynomialGrid(iDim, iOut, iDepth, which_basis, "localp")

    aPoints = grid1.getPoints()
    aVals = np.empty([aPoints.shape[0], 1])
    for iI in range(aPoints.shape[0]):
        aVals[iI] = fn(aPoints[iI][0],aPoints[iI][1],fn_type) 
        #math.cos(2 * math.pi * w1 + c_1*aPnts[iI][0] + c_2*aPnts[iI][1]) 
    grid1.loadNeededPoints(aVals)


    #refinement level
    for iK in range(refinement_level):
        grid1.setSurplusRefinement(fTol, -1, "fds")   #also use fds, or other rules
        aPoints = grid1.getNeededPoints()
        aVals = np.empty([aPoints.shape[0], 1])
        for iI in range(aPoints.shape[0]):
            aVals[iI] = fn(aPoints[iI][0],aPoints[iI][1],fn_type)
            #math.cos(2 * math.pi * w1 + c_1*aPnts[iI][0] + c_2*aPnts[iI][1]) 
        grid1.loadNeededPoints(aVals)

        aRes = grid1.evaluateBatch(aPnts)
        fError1 = max(np.fabs(aRes[:,0] - aTres)) #fabs -> returns absolute values

        print(" {0:9d} {1:9d}  {2:1.2e}".format(iK+1, grid1.getNumPoints(), fError1))


    grid2 = TasmanianSG.TasmanianSparseGrid()
    grid2.makeLocalPolynomialGrid(iDim, iOut, refinement_level+iDepth, which_basis, "localp")
    a = grid2.getNumPoints()

    print("\n-------------------------------------------------------------------------------------------------")
    print( "   a fix sparse grid of level ", refinement_level+iDepth, " would consist of " ,a, " points")
    print("\n-------------------------------------------------------------------------------------------------\n")
    return aPnts, aTres, aPoints, aVals
```

```python
fn_type = 1 

aPnts, aTres, aPoints, aVals = adaptive_sparse_grid(test_fn,fn_type,2,5)

fig = plt.figure()
ax = fig.add_subplot(221,projection='3d')
ax.scatter(aPnts[:,0],aPnts[:,1],aTres[:],c='r',marker='o')
plt.title('Random Grid, Fn 1')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(222,projection='3d')
ax.scatter(aPoints[:,0],aPoints[:,1],aVals[:],c='r',marker='o')
plt.title('Adaptive Sparse Grid, Fn 1')
plt.show()

fn_type = 2 

aPnts, aTres, aPoints, aVals = adaptive_sparse_grid(test_fn,fn_type,3,5)

fig = plt.figure()
ax = fig.add_subplot(223,projection='3d')
ax.scatter(aPnts[:,0],aPnts[:,1],aTres[:],c='r',marker='o')
plt.title('Random Grid, Fn 2')
plt.show()

fig = plt.figure()
ax = fig.add_subplot(223,projection='3d')
ax.scatter(aPoints[:,0],aPoints[:,1],aVals[:],c='r',marker='o')
plt.title('Adaptive Sparse Grid, Fn 2')
plt.show()

```
